```python
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 28

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
def plot_misclassification_errors(dataframes, labels, limits):
    
    plt.figure(figsize=(16,10))
    
    colors = ['orange', 'green', 'blue', 'red']
    
    for label, color, dataframe in zip(labels, colors, dataframes):
        x_values = sorted(dataframe['output_dim'].unique())[9:]

        y_means = []
        y_stds = []

        for x_val in x_values:
            ys = (1 - dataframe.query('output_dim == "{}"'.format(x_val))['test_score'])
            y_means.append(ys.mean())
            y_stds.append(ys.std())
        print(label, y_means)
        
        # plt.errorbar(x_values, y_means, y_stds, linestyle='dotted', marker='^', capsize=3, label=label, color=color, linewidth=4)
        plt.plot(x_values, y_means, linestyle='dotted', marker='^', label=label, color=color, linewidth=4, markersize=10, markeredgewidth=4)
    
    limit_labels = ['Linear Kernel', 'OPU Kernel (m=2)', 'OPU Kernel (m=4)', 'RBF Kernel']
    colors = ['grey', 'green', 'blue', 'red']
    for i, limit in enumerate(limits):
        if limit is not None:
            color = colors[i]
            plt.axhline(limit, label=limit_labels[i] + ' ({})'.format(limit), color=color, linewidth=4)
        
    plt.legend(loc='upper right')
    plt.xlim((0, 10000))
    # plt.xscale('log')
    # plt.yscale('log')
    plt.xlabel(r'Projection Dimension $D$', labelpad=20)
    plt.ylabel('Classification Error')
    plt.title('Kernel Ridge Regression on Fashion MNIST')
    
    plt.savefig('figures/fashion_mnist.pdf', format='pdf')
    # plt.ylim((0.075, 0.25))
    plt.show()
```
